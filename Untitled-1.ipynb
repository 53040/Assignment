```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Replace this with your actual failure cycle data
failure_cycles = [1630,1705,1820,1490,1560,1595,1761,1880,1772,1980,2040,1580,2040,1965,2011,1949,1510,2210,1810,2101,1820]


# Normalize the data
normalized_cycles = np.array(failure_cycles) / max(failure_cycles)

# Fit the Weibull distribution
shape, loc, scale = weibull_min.fit(normalized_cycles, floc=0)

# Calculate the fitted Weibull distribution
x = np.linspace(min(normalized_cycles), max(normalized_cycles), 1000)
y = weibull_min.pdf(x, shape, loc, scale)

# Plot the data points
plt.scatter(normalized_cycles, weibull_min.pdf(normalized_cycles, shape, loc, scale), color='red', label='Data')

# Plot the fitted Weibull distribution line
plt.plot(x, y, label=f'Weibull Distribution\n(shape={shape:.2f}, scale={scale * max(failure_cycles):.2f})')

plt.xlabel('Normalized Failure Cycles')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Weibull Distribution Fit to Failure Cycle Data')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Replace this with your actual failure cycle data

failure_cycles = [1630,1705,1820,1490,1560,1595,1761,1880,1772,1980,2040,1580,2040,1965,2011,1949,1510,2210,1810,2101,1820]

# Fit the Weibull distribution
shape, loc, scale = weibull_min.fit(failure_cycles, floc=0)

# Calculate the fitted Weibull distribution
x = np.linspace(500, 3000, 100)
y = weibull_min.cdf(x, shape, loc, scale)

# Plot the data points
plt.scatter(failure_cycles, weibull_min.cdf(failure_cycles, shape, loc, scale), color='red', label='Data')

# Plot the fitted Weibull distribution line
plt.plot(x, y, label=f'Weibull Distribution\n(shape={shape:.2f}, scale={scale:.2f})')

plt.xlabel('Failure Cycles')
plt.ylabel('Cumulative Probability of Failure')
plt.ylim(0.001, 0.999)
plt.legend()
plt.title('Weibull Distribution Fit to Failure Cycle Data')
plt.show()
```
